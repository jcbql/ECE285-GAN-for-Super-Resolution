```python
from __future__ import print_function
import numpy as np
import torch
import matplotlib.pyplot as plt
import os
import cv2
```

```python
directory = "Caltech256/256_ObjectCategories/"
ldir = os.listdir(directory)
path_bad = []
path = []
for category in os.listdir(directory):
    for filename in os.listdir(directory+category):
        if not filename.endswith('.jpg'): # Some files have no extension for some reason
            path_bad.append(category+'/'+filename)
        else:
            path.append(category+'/'+filename)
# print(cv2.imread(directory+path[100]).shape)
# print(cv2.imread(directory+path[101]).shape)
# im = cv2.imread(directory+path[100])[:,:,::-1] # Change from 'BGR' to 'RGB'
# plt.imshow(im)
# plt.show()
print(path_bad)
print("Why are these 2 files in the dataset??")
```

```python
# cv2.imread(directory+path[30605]).shape[0:2]
test = np.empty([len(path),2])
for i in range(len(path)):
    test[i] = cv2.imread(directory+path[i]).shape[:2]
```

```python
# What is the smallest image size? And which is it?
b = int(np.floor(np.argmin(test)/2.0))
print(b)
print(test[b,:])
```

```python
small=np.where(test<128)[0] # How many images are too small to be used with 128x128 discriminator?
print(len(small),len(path))
print(np.array(path)[small])
```
